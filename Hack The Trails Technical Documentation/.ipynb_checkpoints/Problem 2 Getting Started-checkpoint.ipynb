## Simulating Trail Traffic

This notebook is meant to help you get started on simulating traffic on the trails. It provides some useful meta-data which you can use to begin making such a simulation. This data is available via OSM. As an additional piece of information, you may want to know the locations of the 4 main sensors on the multi-sensor trail. Those locations are as follows:

|Trail Name|Setup Date|Coordinates|In Direction|Out Direction|Notes|
|----------|----------|-----------|------------|-------------|-----|
|Butler Trail - Crenshaw Bridge PC (Urban Trail)|11/30/2015|30.27452 / -97.77122|NORTH|SOUTH|Main Trail (older)|
|Butler Trail - South Lamar PC (Urban Trail)|03/25/2018|30.26430 / -97.75593|EAST|WEST|None|
|Butler Trail - North Congress PC (Urban Trail)|03/25/2018|30.26239 / -97.74426|EAST|WEST|None|
|Butler Trail - Longhorn Dam PC (Urban Trail)|03/25/2018|30.24972 / -97.71372|NORTH|SOUTH|Partial Data|

```python
import numpy as np
import osmnx as ox
import matplotlib.pyplot as plt
from osmnx.core import graph_from_point

%matplotlib inline

# We could get the entire austin Walking map
# G = ox.graph_from_place('Austin, Texas, USA', network_type='walk')

# But instead, we'll retrieve just the walking paths in the area of the primary trail which is bounded by:
G = ox.graph_from_bbox(30.28, 30.24, -97.70 , -97.80, network_type='walk', custom_filter='["motor_vehicle"~"no"]')

# We can plot this and our coordinates above to get the simulation area
fig, ax = ox.plot_graph(G, show=False, close=False)
X, Y = [30.27452, 30.26430, 30.26239, 30.24972], [-97.77122, -97.75593, -97.74426, -97.71372]
ax.scatter(Y, X, c="red")
plt.autoscale()
plt.show()
```
